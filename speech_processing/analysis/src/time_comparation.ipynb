```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def analize_time(file_name: str):
    # Load data
    data = np.loadtxt(f'../data/time_to_finding_f0/{file_name}.txt', dtype=float)
    df = pd.DataFrame(data, columns=['time'])

    # Tính giá trị trung bình và độ lệch chuẩn
    avg = np.mean(data, axis=0)
    std = np.std(data, axis=0)

    print(f'Trung bình (Mean): {avg}')
    print(f'Độ lệch chuẩn (Std Dev): {std}')

    # Tạo biểu đồ violin plot
    plt.figure(figsize=(8, 6))  # Điều chỉnh kích thước của đồ thị
    sns.violinplot(y='time', data=df, width=0.5, color="skyblue")

    # Thêm đường biểu diễn giá trị trung bình
    plt.axhline(avg, color='r', linestyle='--', label=f'Mean = {avg:.2f}')
    
    # Thêm đường biểu diễn độ lệch chuẩn trên và dưới
    plt.axhline(avg + std, color='g', linestyle=':', label=f'Std Dev = {std:.2f}')
    plt.axhline(avg - std, color='g', linestyle=':')

    # Gán tiêu đề và nhãn cho trục
    plt.title('Cepstrum-based Pitch Detection Time Analysis', fontsize=14)
    plt.xlabel('Cepstrum', fontsize=12)
    plt.ylabel('Time (seconds)', fontsize=12)

    # Hiển thị chú thích (legend)
    plt.legend(loc='best')

    # Hiển thị biểu đồ
    plt.show()
```

```python
analize_time('cepstrum')
```
